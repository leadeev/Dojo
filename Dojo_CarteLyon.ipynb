## Challenge 3 - Cartographie

```python
# Affichez les adresses de la rue Delandine, triées par latitude
# Mettez la table correspondante dans une variable "delandine"
rq1 = base_rhone['nom_voie'] == 'Rue Delandine'
delandine = base_rhone[rq1].sort_values(by = 'lat', ascending = False)
delandine
```

```python
# Exécutez cette cellule et demandez-vous pourquoi la ligne a une tête louche
fig, ax = plt.subplots()
ax.plot(delandine.lon, delandine.lat)
fig.show()
#
```

```python
# Répartissez les adresses de la bonne façon en complétant la condition, puis exécutez de sorte à obtenir un truc plus cohérent
condition = delandine["numero"] % 2 == 0

fig, ax = plt.subplots()
ax.plot(delandine[condition].lon, delandine[condition].lat)
ax.plot(delandine[~condition].lon, delandine[~condition].lat)
fig.show()
```
